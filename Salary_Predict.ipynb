```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
dt=DecisionTreeRegressor()
lr=LinearRegression()
lr.fit(x_train,y_train)
dt.fit(x_train,y_train)
rf.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred1=dt.predict(x_test)
y_pred2=rf.predict(x_test)
print(f"Linear Regression {y_pred}")
print(f"Decision Tree Regression {y_pred1}")
print(f" Random Forest Regression {y_pred2}")
```

```python
from sklearn.metrics import r2_score, accuracy_score,mean_absolute_error,mean_squared_error
l1=mean_absolute_error(y_test,y_pred)
l2=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(l1)
print(l2)
print(r2)
l1=mean_absolute_error(y_test,y_pred1)
l2=mean_squared_error(y_test,y_pred1)
r2=r2_score(y_test,y_pred1)
print(l1)
print(l2)
print(r2)
l1=mean_absolute_error(y_test,y_pred2)
l2=mean_squared_error(y_test,y_pred2)
r2=r2_score(y_test,y_pred2)
print(l1)
print(l2)
print(r2)
```
